```python
import numpy as np
import matplotlib.pyplot as plt
import random
import seaborn as sns
import matplotlib.patches as patches
```

```python
def coinflip(P,Q,D):
    r = P+Q   
    prob = []
    for loop in range(200):
        X = []
        for k in range(100):
            x = np.empty((r,D))
            for i in range(r):
                for j in range(D):
                    flip = random.randint(0,1)
                    x[i][j] = flip
            #print(x)
            X.append(x)
            #print(X)
        nsame = 0
        for m in X:
            for n in range(D):
                if m[0][0]!=m[1][0]:
                    nsame += 1
        prob.append(nsame/100)
    prob_mean = np.mean(prob)
    #print(prob)
    plt.axvline(x=prob_mean, linestyle=':',c = 'r', linewidth=2,label = 'probability_mean: %0.4f' % prob_mean)
    plt.legend(loc = 4)
    plt.text(.58, 1.2, 'Theoretical: 0.5', dict(size=10))
    plt.title ('1.(a) Probability Kernel Density Curve')
    sns.distplot(prob,rug=True,hist=False)
```
